```python
import os
import pandas as pd
```

```python
root = '../../download_imgs/sample_output/all/1_50'
d = {
    'uuid': [],
    'path': []
}
for img in os.listdir('./1_50'):
    if img != '.DS_Store':
        uuid = img.split('.')[0]
        path = root + '/' + img
        d['uuid'].append(uuid)
        d['path'].append(path)
df = pd.DataFrame.from_dict(d)
df.info()
```
